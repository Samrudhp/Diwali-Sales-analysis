```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Diwali-Sales-Data.csv',encoding='latin1')
```

```python
data.head()
```

```python
data.info()
```

```python
pd.isnull(data)
```

```python
data.shape
```

```python
#drop unrealed/blank columns
data.drop(['Status','unnamed1'] ,axis = 1, inplace= True)
```

```python
data.shape
```

```python
pd.isnull(data).sum()
```

```python
#dropping the null values
data.dropna(inplace = True)
```

```python
#data type changing
data['Amount'] = data['Amount'].astype('int')
```

```python
data['Amount'].dtypes
```

```python
data[['Amount','Orders','Age']].describe()
```

***Exploratory Data Analysis***


**Gender**

```python
ax = sns.countplot(x = 'Gender',data = data)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#plotting plot for gender v/s total Amount
gender_sales = data.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending = False)
sns.barplot(x = 'Gender',y= 'Amount',data = gender_sales)
```

from above graphs we see that most of the buyers are females and purchasing power of females ate greater than men

**Age**

```python
ax = sns.countplot(data = data,x = 'Age Group', hue ='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = data.groupby(['Age Group','Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
sns.barplot(x = 'Age Group',y = 'Amount',data = sales_age ,hue = 'Gender')
```

from above graphs we can see that most of buyers are of age group between 26-35 yrs female

**State**

```python
#total orders from top 10 states
sales_state = data.groupby(['State'],as_index = False)['Orders'].sum().sort_values(by='Orders',ascending = False)
sns.set(rc = {'figure.figsize':(20,5)})
sns.barplot(data = sales_state,x='State' ,y='Orders')
```

```python
#total orders from top 10 states
sales_state = data.groupby(['State'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending = False)
sns.set(rc = {'figure.figsize':(20,5)})
sns.barplot(data = sales_state,x='State' ,y='Amount')
```

from above graphs we can see that most of the orders & total sales/amount are from UP,Maharashtra and Karnataka

**Marital Status**

```python
ax = sns.countplot( data = data,x = 'Marital_Status',)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_data = data.groupby(['Marital_Status','Gender'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending= False)
sns.set(rc = {'figure.figsize':(6,5)})
sns.barplot(data = sales_data, x = 'Marital_Status',y= 'Amount',hue = 'Gender')
plt.show()
```

From above graphs we can see that most of the buyers are married(women) and they have high purchasing power

**Occupation**

```python

sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = data,x = 'Occupation')
```

```python
sales_data = data.groupby(['Occupation'],as_index= False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data = sales_data,y='Amount',x='Occupation')
```

From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector

**Product Category**

```python
sns.countplot(data = data,x='Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = data.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

**Conclusion**



Married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category
